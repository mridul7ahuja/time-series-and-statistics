```python
import pandas as pd
import math
from scipy import stats
import scipy
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
norm = np.random.normal(0,1,50)
laplace = np.random.laplace(0,1,50)
```

```python
expon = np.random.expon()
```

```python
xarr = []
for x in norm:
    xarr.append(round(x,3))
print(xarr)
xarr = []
for x in laplace:
    xarr.append(round(x,3))
print(xarr)
```

```python
xarrnorm = []
for num in norm:
    xarrnorm.append(num)
meannorm = 0
for num in xarrnorm:
    meannorm += num
meannorm = meannorm/50
print("Mean Norm = " + str(meannorm))
```

```python
xarrlaplace = []
for num in laplace:
    xarrlaplace.append(num)
meanlap = 0
for num in xarrlaplace:
    meanlap += num
meanlap = meanlap/50
print("Mean Lap = " + str(meanlap))
```

```python
xarrnorm.sort()
print("Median Norm = " + str((xarrnorm[24] + xarrnorm[25])/2))
print(xarrnorm)
```

```python
xarrlaplace.sort()
print("Median Lap = " + str((xarrlaplace[24] + xarrlaplace[25])/2))
```

```python
stddevnorm = 0
for num in xarrnorm:
    stddevnorm += (num - meannorm)*(num - meannorm)
stddevnorm = stddevnorm/49
stddevnorm = math.sqrt(stddevnorm)
print("Sample Standard Deviation Norm = " + str(stddevnorm))
```

```python
stddevlap = 0
for num in xarrlaplace:
    stddevlap += (num - meanlap)*(num - meanlap)
stddevlap = stddevlap/49
stddevlap = math.sqrt(stddevlap)
print("Sample Standard Deviation Lap = " + str(stddevlap))
```

```python
coeffVarnorm = stddevnorm/meannorm
print("Coefficient of Variation norm = " + str(coeffVarnorm))
```

```python
coeffVarlap = stddevlap/meanlap
print("Coefficient of Variation lap = " + str(coeffVarlap))
```

```python
skewnorm = 0
for num in xarrnorm:
    skewnorm += (num - meannorm)**3
skewnorm = skewnorm/50
coeffSkewnorm = skewnorm/stddevnorm**3
print("Coefficient of Skewness = " + str(coeffSkewnorm))
```

```python
skewlap = 0
for num in xarrlaplace:
    skewlap += (num - meanlap)**3
skewlap = skewlap/50
coeffSkewlap = skewlap/stddevlap**3
print("Coefficient of Skewness = " + str(coeffSkewlap))
```

```python
kurtosisnorm = 0
for num in xarrnorm:
    kurtosisnorm += (num - meannorm)**4
kurtosisnorm = kurtosisnorm/50
kurtosisnorm = kurtosisnorm/stddevnorm**4
print("Coefficient of Kurtosis = " + str(kurtosisnorm))
```

```python
kurtosislap = 0
for num in xarrlaplace:
    kurtosislap += (num - meanlap)**4
kurtosislap = kurtosislap/50
kurtosislap = kurtosislap/stddevlap**4
print("Coefficient of Kurtosis = " + str(kurtosislap))
```

```python
df = pd.DataFrame({'mesokurtic_Normal':norm,'leptokurtic_Laplace':laplace})
df.describe()
```

```python
df = (df - df.mean())/df.std()

# plotting the three distributions 
sns.set(rc={'figure.figsize':(12,7)})
sns.distplot(df['mesokurtic_Normal'], color='blue', label='Stock A',hist=False)
sns.distplot(df['leptokurtic_Laplace'], color='green', label='Stock B',hist=False)
plt.title('', fontsize=15)
plt.xlabel('Returns over the Quarters')
plt.ylabel('PDF')
plt.legend();
plt.show();
```

```python
results = pd.DataFrame({'MEAN': df.mean().round(2),'STANDARD DEVIATION':df.std().round(2),\
                        'SKEWNESS':df.skew().round(0),'EXCESS KURTOSIS':df.kurtosis().round(0) })
results
```
